```python
import numpy as np
import matplotlib.pyplot as plt

# Define the gradient computation
def gradient(x, y, w, b):
    y_pred = w * x + b
    dw = -2 * np.mean(x * (y - y_pred))
    db = -2 * np.mean(y - y_pred)
    return dw, db

# Generate sample dataset
np.random.seed(0)
x = np.linspace(0, 10, 100)
y = 2 * x + 1 + np.random.normal(0, 1, 100)

# Define a range of values for w and b
w_values = np.linspace(0, 4, 100)
b_values = np.linspace(-2, 4, 100)

# Calculate the gradients for different w values while holding b constant
grad_w = [gradient(x, y, w, 1)[0] for w in w_values]

# Calculate the gradients for different b values while holding w constant
grad_b = [gradient(x, y, 2, b)[1] for b in b_values]

# Create two subplots
plt.figure(figsize=(12, 5))

# Plot 1: Gradient vs w (holding b constant)
plt.subplot(1, 2, 1)
plt.plot(w_values, grad_w, label='Gradient vs w (b=1)', color='blue')
plt.title("Gradient vs w")
plt.xlabel("w")
plt.ylabel("Gradient")
plt.axhline(y=0, color='red', linestyle='--', label='Gradient=0')
plt.legend()

# Plot 2: Gradient vs b (holding w constant)
plt.subplot(1, 2, 2)
plt.plot(b_values, grad_b, label='Gradient vs b (w=2)', color='green')
plt.title("Gradient vs b")
plt.xlabel("b")
plt.ylabel("Gradient")
plt.axhline(y=0, color='red', linestyle='--', label='Gradient=0')
plt.legend()

# Adjust spacing between subplots
plt.tight_layout()

# Show the plots
plt.show()
```
